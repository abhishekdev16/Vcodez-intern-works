```python
import pandas as pd

# Replace 'your_file.csv' with your actual file name
df = pd.read_csv('Mental Health Dataset.csv')

# Now you can check for missing values
df.isnull().sum()
```

```python
# Remove all rows that contain at least one null value
df = df.dropna()

```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```
